```python
temp1 = pd.read_csv('C:/Users/BAUM/Desktop/multicampus/TIL/data/200801_201012_주민등록인구기타현황(출생등록)_birth.csv', encoding='cp949')
temp2 = pd.read_csv(
    'C:/Users/BAUM/Desktop/multicampus/TIL/data/201101_201312_주민등록인구기타현황(출생등록)_birth.csv', encoding='cp949')
temp3 = pd.read_csv(
    'C:/Users/BAUM/Desktop/multicampus/TIL/data/201401_201612_주민등록인구기타현황(출생등록)_birth.csv', encoding='cp949')
temp4 = pd.read_csv(
    'C:/Users/BAUM/Desktop/multicampus/TIL/data/201701_201912_주민등록인구기타현황(출생등록)_birth.csv', encoding='cp949')
temp5 = pd.read_csv(
    'C:/Users/BAUM/Desktop/multicampus/TIL/data/202001_202111_주민등록인구기타현황(출생등록)_birth.csv', encoding='cp949')
```

```python
def make(temp):
    temp = temp.iloc[:, 0::3]
    temp.columns = temp.columns.str.replace('_계', '')
    year = temp.columns[1:].str.split('년').str[0]
    month = temp.columns[1:].str.split('년').str[1].str.split('월').str[0]
    date = year + '-' + month
    date = date.insert(0, '행정구역')
    temp.columns = date
    temp['행정구역'] = temp['행정구역'].str.split().str[0]
    for i in temp.columns[1:]:
        temp[i] = temp[i].str.replace(',', '')
        temp[i] = temp[i].fillna('0').astype(int)
    temp = temp.drop_duplicates(['행정구역'])
    temp = temp.set_index('행정구역')

    return temp.T
```

```python
temp1 = make(temp1)
temp2 = make(temp2)
temp3 = make(temp3)
temp4 = make(temp4)
temp5 = make(temp5)
```

```python
df = pd.concat([temp1, temp2])
df = pd.concat([df, temp3])
df = pd.concat([df, temp4])
df = pd.concat([df, temp5])
```
