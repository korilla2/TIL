```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('indeed_job_dataset.csv')
df.head()
```

```python
df.columns
```

```python
df['Location']
```

```python
df_location = df.groupby('Location')
df_location.count().head()
```

```python
plt.figure(figsize=(20,10))
ax = df_location.count()['Job_Title'].sort_values(ascending=False)
ax = ax.plot(kind='bar', color=['red', 'coral', 'gold', 'yellow', 'forestgreen',
                                                                             "lime",'aquamarine','teal','darkturquoise','deepskyblue',
                                                                             'midnightblue','slateblue','rebeccapurple','indigo','deeppink',
                                                                             'silver','lightcoral','maroon','tomato',"peru",
                                                                             'bisque','olivedrab','palegreen',"cyan","pink"])

plt.title("United States Data Professional Job Market per State (Remote Positions Included)",fontsize=20)
plt.xlabel("Job Location",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
ax1 = df_location.count()['Job_Title'].sort_values(ascending=False).head(15)
ax1 = ax1.plot(kind="bar",color=['red', 'coral', 'gold', 'yellow', 'forestgreen',
                                                                             "lime",'aquamarine','teal','darkturquoise','deepskyblue',
                                                                             'midnightblue','slateblue','rebeccapurple','indigo','deeppink'])
plt.title("United States Data Professional Job Market in Top 15 States",fontsize=20)
plt.xlabel("Job Location",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```
