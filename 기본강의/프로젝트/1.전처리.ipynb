```python
col_list = ['temp']
for i in range(len(df.columns)):
    col_list.append(df.columns[i])

col_list.append('temp')

basic = pd.DataFrame(columns=col_list, index=[
                     '유동비율', '부채비율', 'ROE', 'ROIC', '영업이익률', '매출액증가율', '자기자본증가율'])
```

```python
basic
```

```python
for i in range(1, len(df.columns)):
    basic.loc['유동비율', df.columns[i]] = (
        (df.loc['유동자산', df.columns[i]] / df.loc['유동부채', df.columns[i]]) * 100)
    basic.loc['부채비율', df.columns[i]] = (
        (df.loc['부채총계', df.columns[i]] / df.loc['자본총계', df.columns[i]]) * 100)
    basic.loc['영업이익률', df.columns[i]] = (
        (df.loc['영업이익(손실)', df.columns[i]] / df.loc['영업수익', df.columns[i]]) * 100)
    basic.loc['매출액증가율', df.columns[i]] = (
        ((df.loc['영업수익', df.columns[i]] - df.loc['영업수익', df.columns[i-1]]) / df.loc['영업수익', df.columns[i-1]]) * 100)
    basic.loc['자기자본증가율', df.columns[i]] = (
        ((df.loc['자본총계', df.columns[i]] - df.loc['자본총계', df.columns[i-1]]) / df.loc['자본총계', df.columns[i-1]]) * 100)
    basic.loc['ROE', df.columns[i]] = ((df.loc['당기순이익(손실)', df.columns[i]] / (
        (df.loc['자본총계', df.columns[i-1]] + df.loc['자본총계', df.columns[i]])/2)) * 100)
    basic.loc['ROIC', df.columns[i]] = (((df.loc['영업이익(손실)', df.columns[i]] - df.loc['법인세비용', df.columns[i]]) / (
        (df.loc['자산총계', df.columns[i]] + df.loc['자산총계', df.columns[i]])/2)) * 100)
```
